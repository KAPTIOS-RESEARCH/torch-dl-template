```python
from src.models.vae.vanilla import VanillaVAE

model = VanillaVAE(3, 100)
```

```python
import torch
from torchvision.utils import make_grid
from torchvision import transforms

image = torch.ones((10, 3, 64, 64))
recon = model.generate(image)
```

```python
generated_images = recon.detach().cpu().clamp(0, 1)
grid = make_grid(generated_images, nrow=4)
grid_image = transforms.ToPILImage()(grid)
```
